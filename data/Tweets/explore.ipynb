```python
import numpy as np
import pandas as pd
import os
import ast
import traceback
import psutil
import matplotlib.pyplot as plt
```

## Loading them in

```python
def printmem():
    memory = psutil.virtual_memory()
    print(f"Available Memory: {memory.available / (1024 ** 3)} gb")
```

```python
folder_path = './csv/'

L = os.listdir(folder_path)
for file in L:
    if '.csv' not in file:
        L.remove(file)
```

```python
df = pd.DataFrame()
df2 = pd.DataFrame()
```

```python
for file in L:
    print(f'now adding {file}')
    
    printmem()
    df2 = pd.read_csv(f'./csv/{file}')
    df = pd.concat([df, df2], ignore_index=True)
```

```python
len(df)
```

```python
df.to_csv('bigbaby.csv', index=False)
```
