```python
def augment(l):
    s = l.copy()
    for string in l:
        plural = ''.join([string, 's'])
        s.append(plural)

        if ' ' in string:
            no_spaces = string.replace(' ', '')
            no_spaces_plural = ''.join([no_spaces, 's'])
            with_hyphens = string.replace(' ', '-')
            with_hyphens_plural = ''.join([with_hyphens, 's'])
            
            s.append(no_spaces)
            s.append(no_spaces_plural)
            s.append(with_hyphens)
            s.append(with_hyphens_plural)    
            
    return set(s)
```

```python
from tqdm.notebook import tqdm_notebook
```

```python
df = pd.read_csv('./bigbabies/bigbaby.csv')
```

```python
def subphrase_in_set(sentence, string_set):
    words = sentence.split()
    n = len(words)
    
    for i in range(n):
        for j in range(i+1, n+1):
            subphrase = ' '.join(words[i:j])
            if subphrase in string_set:
                return True
    return False
```

```python
all_list = []
for i, row in tqdm_notebook(df.iterrows(), total=len(df)):
    row['text'] = row['text'].replace('#', '').lower()
    if subphrase_in_set(row['text'], new_filter):
        all_list.append(row)
    if i % 250000 == 0:
        print(len(all_list))
```
