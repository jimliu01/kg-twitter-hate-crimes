```python
top_to_name = top_info.set_index('Topic')['Name'].to_dict()
```

```python
tweet_dict = df.to_dict(orient='records')
```

```python
all_relations=[]
```

```python
tweet_topic_relations = []
for tweet in tqdm(tweet_dict):
    for i in top_set:
        if tweet[f'{i}'] >= 0.01:
            relation = []
            relation.append((tweet['tweet_id'], 't'))
            relation.append(top_to_name[i])
            relation.append(tweet[f'{i}'])
            relation.append(tweet['date'])
            
            tweet_topic_relations.append(relation)

all_relations = [*all_relations, *tweet_topic_relations]
```

```python
all_relations_df = pd.DataFrame(all_relations, columns=['tweet', 'topic', 'weight', 'date'])
all_relations_df.to_csv('./data/tweet-topic-triples.csv', index=False)
```

### Analyzing graph

```python
import networkx as nx
```

```python
graph = nx.Graph()

# Add edges from the DataFrame
for _, row in all_relations_df.iterrows():
    graph.add_edge(row['tweet'], row['topic'], weight=row['weight'])
```

```python
nx.is_connected(graph)
```

```python
import networkx as nx
import community  # python-louvain library

# Apply the Louvain algorithm to detect communities
partition = community.best_partition(graph)

# Print the number of communities and the mapping of nodes to communities
print("Number of communities:", len(set(partition.values())))
print("Node-community mapping:", partition)
```

```python
import networkx as nx
import community
import matplotlib.pyplot as plt

# Apply the Louvain algorithm to detect communities
partition = community.best_partition(graph)

# Create a dictionary to store node colors based on community
node_colors = {node: partition[node] for node in graph.nodes()}

# Create a list of colors to use for the plot
colors = ['red', 'blue', 'green', 'orange', 'purple', 'yellow', 'cyan', 'magenta']

# Create the plot
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(graph, seed=42)  # Choose a layout for the nodes
for community_id in set(partition.values()):
    nodes_in_community = [node for node, comm_id in partition.items() if comm_id == community_id]
    nx.draw_networkx_nodes(graph, pos, nodelist=nodes_in_community, node_color=colors[community_id], label=f"Community {community_id}")

nx.draw_networkx_edges(graph, pos, alpha=0.5)
nx.draw_networkx_labels(graph, pos, font_size=8)
plt.legend(loc='upper right')
plt.axis('off')
plt.title("Communities in the Graph")
plt.show()
```
